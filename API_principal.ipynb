```python
from voronoiq import WeightedVoronoi
from voronoiq.generators import random_generators

pts, w = random_generators(n=200, seed=42)

wv = WeightedVoronoi(pts, w, mode="multiplicative", resolution=512)
wv.compute()

wv.plot()                        # visualização
wv.plot_distance_field()         # campo de distâncias
wv.plot_comparison()             # 3 modos lado a lado

wv.owner(0.5, 0.5)              # índice do gerador dono do ponto
wv.region_of(0.5, 0.5)         # objeto VoronoiRegion
wv.nearest_generators(0.5, 0.5, k=3)

wv.to_png("out.png")
wv.to_svg("out.svg")
wv.to_csv("out.csv")
```
